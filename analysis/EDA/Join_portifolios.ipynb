## Objetivo


Encontrar features não relevantes para remoção


## Método

1 - Aplicar inner join entre portifolios e verificar a repetição de clientes

2 - Nos clientes dos 3 portifolios identificar features sem variação

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
import pandas_profiling
```

```python
p1 = pd.read_csv('../estaticos_portfolio1.csv')
p2 = pd.read_csv('../clientes2_merge.csv')
p3 = pd.read_csv('../clientes3_merge.csv')
exclude = pd.read_csv('../exclude_variables_generated_from_profiling.csv')
```

```python
exclude_variables = exclude.values.squeeze().tolist()
p1.drop(columns=exclude_variables, inplace= True)
p2.drop(columns=exclude_variables, inplace= True)
p3.drop(columns=exclude_variables, inplace= True)
```

### Método 1
Aplicar Inner join entre protifolios

```python
p1.shape,p2.shape,p3.shape
```

```python
 p1.merge(p2, on = 'id', how='inner')
```

```python
p1.merge(p3, on = 'id', how='inner')
```

```python
p23 = p2.merge(p3.id, on = 'id', how='inner')
p23.drop(columns = ['Unnamed: 0'])
```
